# AUTOMATING PREDICTIONS

```python
def textfit(X,Y,ml_model,nlp_model):
    X_c=nlp_model.fit_transform(X)
    x_train,x_test,y_train,y_test=train_test_split(X_c,y)
    ml=ml_model.fit(x_train,y_train)
    acc=ml_model.score(x_test,y_test)
    print(acc)
    
    w=c.get_feature_names()
    coef=ml.coef_.tolist()[0]
    coef_df=pd.DataFrame({'Coeff':coef,'Words':w})
    coef_df=coef_df.sort_values(['Coeff','Words'],ascending=False)
    coef_df1=coef_df[:10]
    sns.barplot(x='Words',y='Coeff',data=coef_df1)
    print("\n")
    print(coef_df.head(20))
    print("The below 20 words :")
    print("\n")
    print(coef_df.tail(20))
    
```

# RNN BASED

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer()
```

```python
X
```

```python
type(X)
```

```python
X=X[:2000]
```

```python
Y=y[:2000]
```

```python
Y
```

```python
X = vectorizer.fit_transform(X).toarray()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y)
```

```python
features=1
```

```python
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1],features)
```

```python
model_rnn = Sequential()
model_rnn.add(SimpleRNN(units=120, activation="relu", input_shape=(60, features)))
model_rnn.add(Dense(units=1))
# Compiling the model
model_rnn.compile(optimizer="Adagrad", loss="binary_crossentropy")
```

```python
model_rnn.fit(X_train, Y_train, validation_data=(X_test,Y_test),epochs=5, batch_size=32)
```

```python
losses_df=pd.DataFrame(model_rnn.history.history)
```

```python
losses_df.plot(kind="line")
```

```python
def predict_sentiment_rnn(texts):
    text=pd.Series(texts)
    x_array = vectorizer.transform(text).toarray() 
    prediction=model_rnn.predict(x_array)
    if prediction[0]>=0.5:
        return "This is a positive statement"
    else:
        return "This is a negative statement"
```

```python
predict_sentiment_rnn("Not good")
```
